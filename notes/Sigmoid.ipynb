## Sigmoid 微分

使用連鎖律和商法則來對函數 $\frac{1}{1+e^{-x}}$ 進行微分。 

將函數 $\frac{1}{1+e^{-x}}$ 寫成 $(1+e^{-x})^{-1}$ 的形式。

使用連鎖律，將外部函數 $u^{-1}$ 和內部函數 $1+e^{-x}$ 分開處理。

對外部函數 $u^{-1}$ 進行微分，得到 $-1\cdot u^{-2}$。

對內部函數 $1+e^{-x}$ 進行微分。
- 步驟 1： 微分常數 $1$。常數的微分為 $0$。
- 步驟 2： 微分 $e^{-x}$。
    - 使用連鎖律(chain rule)。
    - 令 $u=-x$。
    - 先微分外部函數 $e^{u}$，其微分是 $e^{u}$。
    - 再微分內部函數 $u=-x$，其微分是 $-1$。
    - 將兩者相乘，得到 $e^{u}\times (-1)=-e^{u}$。
    - 將 $u$ 換回 $-x$，所以 $e^{-x}$ 的微分是 $-e^{-x}$
- 步驟 3： 將兩部分的微分結果相加。 $0+(-e^{-x})=-e^{-x}$。 
- 最終結果是 $-e^{-x}$。

將外部函數的微分結果與內部函數的微分結果相乘，得到 $-1\cdot (1+e^{-x})^{-2}\cdot (-e^{-x})$。

化簡結果，得到 $\frac{e^{-x}}{(1+e^{-x})^{2}}$。              

最終答案 $\frac{d}{dx}\left(\frac{1}{1+e^{-x}}\right)=\frac{e^{-x}}{(1+e^{-x})^{2}}$

```python
# sigmoid 微分
def sigmoid_derivative(x):
    # sigmoid = 1 / (1 + np.exp(-x))
    return np.exp(-x) / (1 + np.exp(-x))**2

x = np.linspace(-10, 10, 100)
y = sigmoid_derivative(x)
s = 1 / (1 + np.exp(-x))  # for reference
plt.plot(x, y)
plt.plot(x, s, linestyle='dashed')
# plt.ylim(-3.1, 3.1)
plt.grid()
plt.show()
```
